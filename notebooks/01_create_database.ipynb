```python
import pandas as pd
import numpy as np
import sqlite3
import os

# Load CSV
df = pd.read_csv("../data/diabetes.csv")

# Clean invalid zeros
cols_with_zero = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
df[cols_with_zero] = df[cols_with_zero].replace(0, np.nan)
df[cols_with_zero] = df[cols_with_zero].fillna(df[cols_with_zero].median())

# Ensure data directory exists
os.makedirs("../data", exist_ok=True)

# Create database
conn = sqlite3.connect("../data/diabetes.db")
cursor = conn.cursor()

# Create tables
cursor.execute("""
CREATE TABLE patients (
    patient_id INTEGER PRIMARY KEY AUTOINCREMENT,
    pregnancies INTEGER,
    glucose REAL,
    blood_pressure REAL,
    skin_thickness REAL,
    insulin REAL,
    bmi REAL,
    diabetes_pedigree REAL,
    age INTEGER
)
""")

cursor.execute("""
CREATE TABLE labels (
    patient_id INTEGER,
    outcome INTEGER,
    FOREIGN KEY (patient_id) REFERENCES patients(patient_id)
)
""")

# Insert data
for _, row in df.iterrows():
    cursor.execute("""
        INSERT INTO patients (
            pregnancies, glucose, blood_pressure,
            skin_thickness, insulin, bmi,
            diabetes_pedigree, age
        )
        VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    """, (
        int(row["Pregnancies"]),
        float(row["Glucose"]),
        float(row["BloodPressure"]),
        float(row["SkinThickness"]),
        float(row["Insulin"]),
        float(row["BMI"]),
        float(row["DiabetesPedigreeFunction"]),
        int(row["Age"])
    ))

    pid = cursor.lastrowid
    cursor.execute(
        "INSERT INTO labels (patient_id, outcome) VALUES (?, ?)",
        (pid, int(row["Outcome"]))
    )

conn.commit()

# Verify tables and counts
print(pd.read_sql("SELECT COUNT(*) FROM patients", conn))
print(pd.read_sql("SELECT COUNT(*) FROM labels", conn))

conn.close()
print("DATABASE CREATED SUCCESSFULLY")
```
